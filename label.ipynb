```python
import os
import pandas as pd
import subprocess
import soundfile as sf
from tqdm import tqdm
```

```python
df = pd.read_csv('labels.csv',header=None)
df.rename(columns={0 : 'index', 1 : 'genres'}, inplace= True)
df
```

```python
result_dict = df.set_index('index').to_dict()['genres']
result_dict
```

```python
genress = df['genres'].unique()
```

```python
genress
```

```python
# Run the shell script
for i in genress:
    shell_script = f'''
    #!/bin/bash
    mkdir Data/fma_medium/{i}
    '''
    subprocess.run(["bash", "-c", shell_script])
```

```python
folder = os.listdir('Data/fma_medium/')
folder = [x for x in folder if x.isdigit()]
folder
```

```python
dir = []
for i in folder:
    for audio in os.listdir('Data/fma_medium/' + i):
        id = audio.split('.')[0]
        genre = result_dict[int(id)]
        # dir.append(genre)
        shell_script = f'''
        #!/bin/bash
        mv Data/fma_medium/{i}/{audio} Data/fma_medium/{genre}
        '''
        subprocess.run(["bash", "-c", shell_script])
shell_script
```

```python
from collections import Counter
c  = Counter(dir)
c
```

```python
file_list = []
dir = 'Data/fma_medium/'
for i, (dirpath, dirnames, filenames) in enumerate(os.walk(dir)):
        if dirpath is not dir:
                for file in filenames:
                        file_list.append(os.path.join(dirpath, file))
                        
```

```python
for i in tqdm(file_list):
    try: 
        sf.read(i)
    except:
        shell_script = f'''
        #!/bin/bash
        rm {i}
        '''
        subprocess.run(["bash", "-c", shell_script])
```
