Uneven number of item leads to dataloader loading uneven test set. We need to fix it equally, ideally around 100 tem per set.

```python
def select_random_items(folder_path, num_items):
    all_items = os.listdir(folder_path)
    selected_items = random.sample(all_items, num_items)
    
    return selected_items

def remove_items(folder_path, selected_items):
    all_items = os.listdir(folder_path)
    
    not_selected_items = [item for item in all_items if item not in selected_items]

    for item in not_selected_items:
        item_path = os.path.join(folder_path, item)
        os.remove(item_path)
```

```python
dir = 'Data/fma_medium/'
for genre in ['Country', 'Classical', 'Pop', 'Hip-Hop', 'Rock', 'Jazz']:
    nextdir = dir+genre
    _select = select_random_items(nextdir,100)
    remove_items(nextdir,_select)
```

Then we need to clone some items in blues so that it reaches 100

```python
_numtoclone = 100 - 74
folder_path = 'Data/fma_medium/Blues'
items = select_random_items(folder_path, _numtoclone)
for item in items:
    item_path = os.path.join(folder_path, item)
    clone_path = os.path.join(folder_path, f"clone_{item}")
    shutil.copy(item_path, clone_path)
```
